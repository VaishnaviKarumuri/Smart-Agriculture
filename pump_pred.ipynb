```python
from IPython.core.formatters import PlainTextFormatter
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#importing dataset
df=pd.read_csv('moisture_days.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,1].values
#Splitting the dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.8, random_state=0)
#Fitting simple linear regression to the training set
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train, y_train)
X_test[0]=1000;
X_test[1]=2000;
#predicting test set results
y_pred=regressor.predict(X_test)
days=y_pred[0]-y_pred[1]
print('\n \n \t Water needed after='+str(days)+'days')
#Visualizing the training set results
plt.scatter(X_train, y_train, color='purple')
plt.plot(X_train, regressor.predict(X_train), color='green')
plt.title('Time vs Moisture(Training set)')
plt.xlabel('moisture')
plt.ylabel('time')
plt.show()
#Visualizing the test set results
plt.scatter(X_test, y_test, color='purple')
plt.plot(X_train, regressor.predict(X_train), color='green')
plt.title('Time vs Moisture(Testing set)')
plt.xlabel('moisture')
plt.ylabel('time')
plt.show()
print('\n \n \t Water needed after='+str(days)+'days')
```

DURATION OF IRRIGATION

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Importing dataset
df = pd.read_csv('moisture_time.csv')

# Extracting features (moisture) and target variable (time)
X = df[['Moisture']].values
y = df['time'].values

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=0)

# Fitting linear regression to the training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Making predictions on the entire dataset (for visualization)
y_pred = regressor.predict(X)

# Visualizing the results
plt.scatter(X, y, color='purple', label='Actual Data')
plt.plot(X, y_pred, color='green', label='Linear Regression')
plt.title('Moisture vs Time')
plt.xlabel('Moisture')
plt.ylabel('Time')
plt.legend()
plt.show()

# Assuming you have a specific moisture level for irrigation, let's say 800
moisture_for_irrigation = 800

# Predicting the time for irrigation
irrigation_duration = regressor.predict([[moisture_for_irrigation]])

print(f'Irrigation Duration for Moisture Level {moisture_for_irrigation}: {irrigation_duration[0]:.2f} units of time')
```
